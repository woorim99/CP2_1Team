```python
import sqlite3
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

### 필요한 데이터 추출, view는 1점, cart는 2점, purchase는 3점 부여

```python
cosine_test = pd.read_csv('testview.csv')
cosine_test = cosine_test[['user_id', 'product_id', 'event_type']]
cosine_test['event_type'] = cosine_test['event_type'] + 1
cosine_test
```

### user_id와 product_id 별로 합친 점수

```python
cosine = cosine_test.groupby(['user_id', 'product_id'], as_index=False).sum()
cosine
```

### 제품 별 유사도를 측정하기 위해 pivot table 구성

```python
product_user_cosine = cosine.iloc[:100000].pivot_table('event_type', index = 'product_id', columns='user_id')
product_user_cosine
```

```python
product_user_cosine.fillna(0, inplace = True)
product_user_cosine
```

### 코사인 유사도

```python
item_based_collabor = cosine_similarity(product_user_cosine)
item_based_collabor
```

```python
item_based_collabor = pd.DataFrame(data = item_based_collabor, index = product_user_cosine.index, columns = product_user_cosine.index)
item_based_collabor
```

```python
item_based_collabor.to_csv('cosine_similarity.csv')
```

### 입력한 product_id에 해당하는 제품과 비슷한 5개의 제품 출력(입력된 제품 포함)

```python
def get_item_based_collabor(product_id):
    return item_based_collabor[product_id].sort_values(ascending=False)[:6]
```

### 입력한 product_id에 해당하는 제품과 비슷한 5개의 제품 정보 출력(입력된 제품 포함)

```python
conn = sqlite3.connect('cp2_data.db')
cur = conn.cursor()

query = cur.execute("""SELECT category_code, event_type, brand, price
FROM "2019-Oct"
WHERE category_code != '' AND brand != '' WHERE product_id = {product_id} LIMIT 1""")

cols = [column[0] for column in query.description]

df = pd.DataFrame.from_records(data=query.fetchall(), columns=cols)

conn.close()
```

```python
def get_item_based_collabor(product_id):
    recommend = item_based_collabor[product_id].sort_values(ascending=False)[:6].index
    recommend_list = []
    for i in recommend:
        i = str(i)
        conn = sqlite3.connect('cp2_data.db')
        cur = conn.cursor()

        query = cur.execute("""SELECT category_code, event_type, brand, price
        FROM "2019-Oct"
        WHERE category_code != '' AND brand != '' AND product_id == {} LIMIT 1""".format(i))

        cols = [column[0] for column in query.description]

        df = pd.DataFrame.from_records(data=query.fetchall(), columns=cols)

        conn.close()
        recommend_list.append({'product id': i, 'category code': df['category_code'].iloc[0], 'brand': df['brand'].iloc[0], 'price': df['price'].iloc[0]})
    return recommend_list
```

### 추천시스템

```python
# 제품id를 넣었을 때 유사한 제품 5개 리스트를 반환
def get_item_based_collabor(product_id):
    recommend = item_based_collabor[product_id].sort_values(ascending=False)[:6].index
    recommend_list = []
    for i in recommend:
        i = str(i)
        conn = sqlite3.connect('cp2_data.db')
        cur = conn.cursor()

        query = cur.execute("""SELECT category_code, event_type, brand, price
        FROM "2019-Oct"
        WHERE category_code != '' AND brand != '' AND product_id == {} LIMIT 1""".format(i))

        cols = [column[0] for column in query.description]

        df = pd.DataFrame.from_records(data=query.fetchall(), columns=cols)

        conn.close()
        recommend_list.append({'product id': i, 'category code': df['category_code'].iloc[0], 'brand': df['brand'].iloc[0], 'price': df['price'].iloc[0]})
    return recommend_list

# 고객id를 넣었을 때 추천할 제품 5개 정보를 반환
def recommend_item(user_id):
    final_recommend = []
    recent_product_id = cosine_test[cosine_test['user_id'] == user_id].iloc[-1]['product_id']
    recommend_item = get_item_based_collabor(recent_product_id)[1:]
    for item in recommend_item:
        try:
            idx = test[test['product_id'] == item['product id']].index[0]
            if pred[idx] == 1:
                final_recommend.insert(0, item)
                
        except:
            final_recommend.append(item)
            
    return final_recommend

recommend_item(558670273)
```
